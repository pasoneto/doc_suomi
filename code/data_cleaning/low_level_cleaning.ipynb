```r
setwd("/home/pasoneto/Documents/github/doc_suomi/code")
source("utils.R")
source("data_cook.R")
cat(howto_data)

oi = fread("/home/pasoneto/Documents/github/doc_suomi/data/treated_data/data_low.csv")
oi = split(oi, oi$album_id)

# Getting out albums with outlier number of observations
tamanhos = c()
for(i in 1:length(oi)){
    tamanhos <- c(tamanhos, NROW(oi[[i]]))
}

out <- boxplot.stats(tamanhos)$out
out_ind <- which(tamanhos %in% c(out))

for(i in out_ind){
    oi[[i]] <- NULL
}
ha = dplyr::bind_rows(oi)

ha %>% group_by(album_id, track_id) %>%
    summarise(n_obs = NROW(valence)) -> obs

# Getting tracks with outlier number of samples and removing whole albums
out <- boxplot.stats(obs$n_obs)$out

# removing whole albums
ha = ha %>% filter(!album_id %in% unique(ha[which(obs$n_obs %in% c(out)), ]$album_id))

# removing albums with missing track number
ha %>% group_by(album_id) %>%
    summarise(length = max(track_number), 
              length2 = length(unique(track_number))) %>%
    filter(length != length2) %>% as.list(select(album_id)) -> albums

ha %<>% filter(!album_id %in% c(unlist(albums$album_id)))
```
