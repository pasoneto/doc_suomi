```python
import numpy as np
from PIL import Image
import pandas as pd

#receives data frame, width and length of the output
def parser(data, w, l):
    data = np.array(data, dtype=np.double)
    convert = Image.fromarray(np.array(data, dtype=np.double)).resize((w, l), Image.NEAREST)
    return np.array(convert)

def split(df, group):
    gb = df.groupby(group)
    return [gb.get_group(x) for x in gb.groups]
```

```python
# Processing
dados = pd.read_csv("/home/pasoneto/Documents/Ciência/PhD/data/treatment/sem_duplica_real.csv")[["album_id", 'valence', 'energy', 'loudness', 'tempo']]
dados['album_id'] = list(dados.groupby('album_id').ngroup())
dados = split(dados, 'album_id')
```

```python
oi = [pd.DataFrame(parser(x[['valence', 'energy', 'loudness', 'tempo']], 4, 16)) for x in dados]

#adding back album id
for i in range(len(oi)):
    oi[i]['album_id'] = i
    oi[i]['track_number'] = range(16)
```

```python
f = pd.concat(oi)
```

```python
f.to_csv('upsampled_filtered.csv')
```
